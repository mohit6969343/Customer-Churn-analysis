```python
plt.figure(figsize=(3,3))
sns.countplot(x="SeniorCitizen" , data=df )
plt.title("churn by SeniorCitizen")
plt.show()
```

```python
senior_churn = pd.crosstab(df['SeniorCitizen'], df['Churn'])

# Convert to percentage
senior_churn_percent = senior_churn.div(senior_churn.sum(axis=1), axis=0) * 100

# Plot
ax = senior_churn_percent.plot(kind='bar', stacked=True, figsize=(5,5), color=['#66c2a5', '#fc8d62'])

# Add percentage labels
for idx, row in senior_churn_percent.iterrows():
    cumulative = 0
    for i, val in enumerate(row):
        plt.text(
            x=idx,
            y=cumulative + val / 2,
            s=f'{val:.1f}%',
            ha='center',
            va='center',
            fontsize=9,
            color='black'
        )
        cumulative += val

# Formatting
plt.title("Churn Percentage by Senior Citizen Status")
plt.xlabel("Senior Citizen (0 = No, 1 = Yes)")
plt.ylabel("Percentage")
plt.legend(title='Churn', loc='best')
plt.ylim(0, 100)
plt.tight_layout()
plt.show()
```

```python
#comparative a greater percentage of people in senior citizen have churned
```

```python
cols = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Set up the subplot grid
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(15, 12))
axes = axes.flatten()  # Flatten for easy iteration

# Loop over each column and plot
for i, col in enumerate(cols):
    sns.countplot(x=col, data=df, ax=axes[i], palette='pastel')
    axes[i].set_title(f'Countplot of {col}')
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Count')
    axes[i].tick_params(axis='x', rotation=45)

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
#The majority of customers who do not churn tend to have services like PhoneService, internetServices(particularly DSL) and ONline Security enabled
#ONlinebackup, TechSupport, and Streaming TV churn rates are noticeably higher when these services are not used or are unavailable
```
